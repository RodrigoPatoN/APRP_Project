```python
import pandas as pd
```

```python
data = pd.read_csv("./LoanData_Bondora.csv")

print(data.shape)

data
```

```python
drop_cols = []

for column in data.columns: 
    num_nan = data[column].isna().sum()
    pct = (num_nan/data.shape[0])*100
    
    if pct > 30.0:
        
        drop_cols.append(column)
    
    print(f"{column}: {num_nan} ({pct} %)")
```

As we can see, there are a lot of missing values.

As such, we will drop all column with over 30% missing values.

```python
data = data.drop(drop_cols, axis = 1)

data.shape
```

```python
cols_const = []

for column in data.columns:
    if len(data[column].unique()) == 1:
        print(column)
        cols_const.append(column)

print(len(cols_const))
```

```python
data = data.drop(cols_const, axis = 1)

data.shape
```

We also dropped "ReportAsOfEOD" as that column was constant.

Besides those columns, there are some other columns that provide no usefull information.

We will also drop these columns.

```python
drop_cols_2 = ['LoanId', 'LoanNumber', 'ListedOnUTC', 'BiddingStartedOn',
            'UserName', 'LanguageCode','LoanApplicationStartedDate',
            'LoanApplicationStartedDate','ApplicationSignedHour', 'ApplicationSignedWeekday',
            'ActiveScheduleFirstPaymentReached','ModelVersion','WorseLateCategory',
            'ProbabilityOfDefault', 'ExpectedLoss', 'LossGivenDefault', 'ExpectedReturn',
            'LoanDate', 'MaturityDate_Original' , 'MaturityDate_Last' ,'LastPaymentOn']

data = data.drop(drop_cols_2, axis = 1)

data.shape
```
